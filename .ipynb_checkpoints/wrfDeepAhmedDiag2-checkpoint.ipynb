```python
def get_density(a,bins=np.linspace(-1,0.2,151)):
    print(a.shape)
    return np.histogram(a,bins,density=True)[0]

def get_density_dask_fast(a,bins):
    return dask.array.histogram(a,bins,density=True)[0].compute()


def get_binned_dask_fast(a,b,bins, return_pdf=False):
    idx_bin_inf = max(0,np.argmax(bins>float(a.min()))-1)
    idx_bin_sup = np.argmin(bins<float(a.max()))
    if idx_bin_sup==0 or idx_bin_sup==len(bins)-1:
        bins=bins[idx_bin_inf:]
    else:
        bins=bins[idx_bin_inf:idx_bin_sup+1]
    bin_centers=(bins[1:]+bins[:-1])/2
    
    bin_sums = dask.array.histogram(a,bins,weights=b)[0]
    counts = dask.array.histogram(a,bins,density=False)[0]
    density = dask.array.histogram(a,bins,density=True)[0]
    if return_pdf:
        return bin_centers,density.compute(),(bin_sums/counts).compute()
    else:
        return bin_centers,(bin_sums/counts).compute()
```

```python
import dask.array as da
def get_binned_np(a,bins=[0,1],n=0):
    bin_sums = np.histogram(a[:n],bins,weights=a[n:])[0]
    counts   = np.histogram(a[:n],bins,density=False)[0]
    return bin_sums/counts
def get_binned_along_x(a,b,bins):
    return da.apply_along_axis(get_binned_np, 0, 
                               da.concatenate((a.reshape(a.shape[0]*a.shape[1],a.shape[2]),
                                               b.reshape(a.shape[0]*a.shape[1],a.shape[2]))),
                               bins=bins,
                               n=a.shape[0]*a.shape[1],
                               dtype=np.float64,
                               shape=(len(bins)-1,)
                              ).T.compute()

```

```python
def fit(pred,dep,model='linear'):
    predrange = np.linspace(np.min(pred),np.max(pred),41)
    if model=='linear':
        func = lambda x,a,b : a*x+b
    elif model=='exponential':
        func = lambda x,a,b : np.exp(a*x+b)
    elif model=='softplus':
        func = lambda x,a,b : np.log(1+np.exp(a*x+b))
    popt,_ = curve_fit(func,pred,dep)
    fitrange = func(predrange,*popt)
    residual = dep - func(pred,*popt)
    rss = np.sum(residual**2)
    tss = len(dep)*np.var(dep)
    rsq = 1-rss/tss
    return predrange,fitrange,rsq
func = lambda x,a,b : np.exp(a*x+b)
from scipy.optimize import curve_fit
```

```python
n=107
popt,_ = curve_fit(func,Sim.BLbins[:n],Sim.RAINNC_BLbinned[:n])
fit = lambda x : np.exp(popt[0]*x+popt[1])
plt.plot(Sim.BLbins,Sim.RAINNC_BLbinned)
plt.plot(Sim.BLbins,fit(Sim.BLbins))
plt.ylim(0,150)
```
