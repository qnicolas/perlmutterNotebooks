```python
def calc_w_general(Sim):
    xx ,hx = topographic_profile('cos',A=500)
    xx = xx[1000:3001]
    hx = hx[1000:3001]
    
    z=np.linspace(0,20e3,100)
    p = interp1d(Sim.Z_PL_SNAP,Sim.Z_PL_SNAP.pressure,kind='linear',fill_value='extrapolate')(z)
    N = interp1d(Sim.Z_PL_SNAP,Sim.N_PL_MEAN,kind='linear',fill_value='extrapolate')(z)
    U = 10.*z**0
    w_general = linear_w_generalized(xx,hx,z,U,N)
    return w_general.assign_coords({'pressure':('altitude',p)}).swap_dims({'altitude':'pressure'})

def calc_Tprime(Sim,w):
    z=np.array(w.altitude)*1000
    N = interp1d(Sim.Z_PL_SNAP,Sim.N_PL_MEAN,kind='linear',fill_value='extrapolate')(z)
    U = 10.*z**0
    return w_to_Tprime(w,U,N)
```

```python
w_300_general = calc_w_general(W10S300)
w_296_general = calc_w_general(W10S296)
```

# With diabatic heating

```python
for Sim in W10S296,W10S300:
    Sim.U_PL_MEAN = xr.open_dataarray(Sim.path+'wrf.U_PL.ALLMEAN.nc')
    Sim.RHO_PL_MEAN = Sim.T_PL_MEAN.pressure*100./287./Sim.T_PL_MEAN
    Sim.OMEGA_PL_MEAN = -Sim.RHO_PL_MEAN*9.81*Sim.W_PL_MEAN
    pi = (Sim.T_PL_MEAN.pressure/1e3)**0.286
    Sim.THETA_PL_MEAN = Sim.T_PL_MEAN/pi
    udtdx = -(Sim.U_PL_MEAN*Sim.T_PL_MEAN.differentiate('distance_from_mtn')/1000 * 86400)
    omegadsdp = (pi*Sim.OMEGA_PL_MEAN*Sim.THETA_PL_MEAN.differentiate('pressure')/100 * 86400)
    Qdiab = udtdx+omegadsdp
    Sim.QDIAB_Kps = Qdiab.where((np.abs(Qdiab)<50)).fillna(0.)
    Sim.QDIAB_B = 9.81*(Sim.QDIAB_Kps/Sim.T_PL_MEAN).interp_like(w_300_general).fillna(0.).transpose('distance_from_mtn','pressure')
```

```python

def calc_w_general_forced(Sim,q_diab_b,nomtn=False):
    xx ,hx = topographic_profile('cos',A=500)
    xx = xx[1000:3001]
    hx = hx[1000:3001]
    
    if nomtn:
        hx*=0
    
    z=np.linspace(0,20e3,100)
    p = interp1d(Sim.Z_PL_SNAP,Sim.Z_PL_SNAP.pressure,kind='linear',fill_value='extrapolate')(z)
    N = interp1d(Sim.Z_PL_SNAP,Sim.N_PL_MEAN,kind='linear',fill_value='extrapolate')(z)
    U = 10.*z**0
    w_general = linear_w_generalized_forced(xx,hx,z,U,N,np.array(q_diab_b)/86400)
    return w_general.assign_coords({'pressure':('altitude',p)}).swap_dims({'altitude':'pressure'})
```

## Correlating low-level W with precip

```python
def get_binned_dask_fast(a,b,bins, return_pdf=False):
    idx_bin_inf = max(0,np.argmax(bins>float(a.min()))-1)
    idx_bin_sup = np.argmin(bins<float(a.max()))
    if idx_bin_sup==0 or idx_bin_sup==len(bins)-1:
        bins=bins[idx_bin_inf:]
    else:
        bins=bins[idx_bin_inf:idx_bin_sup+1]
    bin_centers=(bins[1:]+bins[:-1])/2
    
    bin_sums = dask.array.histogram(a,bins,weights=b)[0]
    counts = dask.array.histogram(a,bins,density=False)[0]
    density = dask.array.histogram(a,bins,density=True)[0]
    if return_pdf:
        return bin_centers,density.compute(),(bin_sums/counts).compute()
    else:
        return bin_centers,(bin_sums/counts).compute()
def get_binned_np(a,b,bins, return_pdf=False):
    idx_bin_inf = max(0,np.argmax(bins>float(a.min()))-1)
    idx_bin_sup = np.argmin(bins<float(a.max()))
    if idx_bin_sup==0 or idx_bin_sup==len(bins)-1:
        bins=bins[idx_bin_inf:]
    else:
        bins=bins[idx_bin_inf:idx_bin_sup+1]
    bin_centers=(bins[1:]+bins[:-1])/2
    
    bin_sums = np.histogram(a,bins,weights=b)[0]
    counts   = np.histogram(a,bins,density=False)[0]
    density  = np.histogram(a,bins,density=True)[0]
    if return_pdf:
        return bin_centers,density,(bin_sums/counts)
    else:
        return bin_centers,(bin_sums/counts)
```

# An idealized heat source

```python
x = W10S300.QDIAB_B.distance_from_mtn
p = W10S300.QDIAB_B.pressure
step = lambda a : np.arctan(a/2)/np.pi+0.5
qdiab_b_synthetic = 0.4*(step(x+100)*step(100-x))*step(p-950)
qdiab_b_synthetic2 = qdiab_b_synthetic + 0.4*(step(x+100)*step(-x))*(p<=940.)*np.exp(-(p-950)**2/(2*100**2))
```
