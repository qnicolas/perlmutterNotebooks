```python
g = 9.81
Lv = 2.5e6
cp=1004.
Rd = 287.
Rv = 461.

Tm = 250.
Tt = 200.
```

```python
def Qrad_Jpm3ps(T,p):
    rho = p/(Rd*T)
    cst = -rho*cp*1/86400 # 1 K/day
    return (T>=Tm)*cst + (T>Tt)*(T<Tm)*0.5*(1+np.cos(np.pi*(Tm-T)/(Tm-Tt)))*cst + (T<=Tt)*1e-6
```

```python
def calc_Gamma_m(T,qs):
    return g/cp * (1+qs*Lv/(Rd*T))/(1+qs*Lv**2/(cp*Rv*T**2))

def calc_gamma_m1(T,Gamma_m):
    """This is the Singh version. Pretty sure it contains a mistake"""
    return Lv/(Rv*T**2)*(Gamma_m-g/(Rd*T))

def calc_gamma_m2(T,Gamma_m):
    return Lv/(Rv*T**2)*Gamma_m - g/(Rd*T)
    
def calc_b(T,qs):
    return Lv**2*qs/(cp*Rv*T**2+Lv**2*qs)

class ZPB():
    def __init__(self,zz,ww,Tref = 243.8, pref=370e2, zref = 8e3, epsilon = 0.6e-3, mu=1.5, typ=2):
        self.zz = zz
        self.ww = ww
        self.Tref = Tref
        self.pref = pref
        self.zref = zref
        self.epsilon=epsilon
        self.delta=epsilon
        self.mu=mu
        
        if typ==1:
            self.calc_gamma_m = calc_gamma_m1
        elif typ==2:
            self.calc_gamma_m = calc_gamma_m2
        
    def calc_RH(self,T,p,qs,Gamma_m,w):
        gamma_m = self.calc_gamma_m(T,Gamma_m)
        b = calc_b(T,qs)
        epsilon_hat = self.epsilon/gamma_m
        delta_hat = self.delta/gamma_m
        
        rho = p/(Rd*T)
        w_hat = Lv*qs*gamma_m/Qrad_Jpm3ps(T,p)*rho*w
        
        a3 = epsilon_hat*b*self.mu/(1+self.mu)*w_hat
        a2 = epsilon_hat*b/delta_hat/(1+self.mu) + (epsilon_hat*(1-b) - (1+2*epsilon_hat*b)*(self.mu/(1+self.mu)))*w_hat
        a1 = -(1 + (1+epsilon_hat*b)/(delta_hat*(1+self.mu))) + (epsilon_hat*(b-2) + (1+epsilon_hat*b)*(1+2*self.mu)/(1+self.mu))*w_hat
        a0 = 1 + (epsilon_hat*(1-b) - 1)*w_hat
    
        roots = np.roots(np.array([a3,a2,a1,a0]))
        crit = (np.real(roots)>=0)*(np.real(roots)<=1)*(np.abs(np.imag(roots))<1e-5)
        if np.sum(crit)==0:
            raise ValueError('issue in RH root, roots = ',roots)
    
        candidate_idx = np.argmax(crit)
        RH = np.real(roots[candidate_idx])
        if (RH < 0) or (RH>1):
            raise ValueError(RH)
        return RH
    
    def calc_Gamma(self,T,p,w):
        qs = qsat(T,p/100)
        Gamma_m = calc_Gamma_m(T,qs)
        RH = self.calc_RH(T,p,qs,Gamma_m,w)
        return Gamma_m + self.epsilon*Lv*qs*(1-RH)/(cp+qs*Lv**2/(Rv*T**2))
        
    def dTPdz(self,TP,z):
        w = np.interp(z,self.zz,self.ww)
        return np.array([-self.calc_Gamma(TP[0],TP[1],w),-TP[1]*g/Rd/TP[0]])

    def integrate(self):
        zz_above = self.zz[self.zz>=self.zref]
        zz_below = self.zz[(self.zz<=self.zref)&(self.zz>=500.)][::-1]
        zz_blayer = self.zz[(self.zz<=500.)]

        TP_above = odeint(self.dTPdz,np.array([self.Tref,self.pref]),zz_above)
        TP_below = odeint(self.dTPdz,np.array([self.Tref,self.pref]),zz_below)
        
        TT_blayer = TP_below[-1,0]-g/cp*(zz_blayer-500.)
        TT = np.concatenate((TT_blayer[:-1],TP_below[::-1,0],TP_above[1:,0]))
        TT[TT<Tt] = Tt
        self.TT = TT
        
        integ = cumtrapz(g/Rd/TT_blayer,zz_blayer,initial=0)
        pp_blayer = TP_below[-1,1]*np.exp(integ[-1]-integ[:-1])
        self.pp = np.concatenate((pp_blayer,TP_below[::-1,1],TP_above[1:,1]))
#
        #qsat_ = qsat(self.TT,self.pp/100)
        #Gamma_m_ = calc_Gamma_m(self.TT,qsat_)
        #self.RH_ = np.array([self.calc_RH(self.TT[i],self.pp[i],qsat_[i],Gamma_m_[i],self.ww[i]) for i in range(len(self.zz))])
        #self.qq = self.RH_*qsat_
        self.rho_ = self.pp/Rd/self.TT
    
    def calc_r(self,T,p,w):
        qs = qsat(T,p/100)
        Gamma_m = calc_Gamma_m(T,qs)
        rho = p/Rd/T
        
        RH = self.calc_RH(T,p,qs,Gamma_m,w)  
        b = calc_b(T,qs)
        gamma_m = self.calc_gamma_m(T,Gamma_m)
        
        r = self.delta*(1+self.mu)*(1-RH)/(RH*(gamma_m+self.epsilon*b*(1-RH)))
        return r
    
    def calc_precip(self):
        if not hasattr(self,'TT'):
            raise ValueError('You need to call integrate() first.')
        r_ = np.array([self.calc_r(self.TT[i],self.pp[i],self.ww[i]) for i in range(len(self.zz))])
        
        qsat_ = qsat(self.TT,self.pp/100)
        Gamma_m_ = calc_Gamma_m(self.TT,qsat_)
        RH_ = np.array([self.calc_RH(self.TT[i],self.pp[i],qsat_[i],Gamma_m_[i],self.ww[i]) for i in range(len(self.zz))])
        self.r_ = r_
        snet = -1/r_ * ( Qrad_Jpm3ps(self.TT,self.pp)/Lv - self.mu*self.delta*(self.pp/Rd/self.TT)*self.ww*(1-RH_)*qsat_)
        snet[self.TT<=Tt] = 0.
        
        self.snet = snet
        self.precip = np.trapz(self.snet,self.zz)
        
    def calc_Q(self):
        if not hasattr(self,'snet'):
            raise ValueError('You need to call calc_precip() first.')    
        self.Q = Lv*self.snet + Qrad_Jpm3ps(self.TT,self.pp)
    
```

# DGW coupling

```python
from orographicPrecipitation.observationsAndERA5.linearWaveModels import second_derivative_matrix
def dgw_mode(z_,rhs_):
    z = z_[z_<11e3]
    rhs = rhs_[z_<11e3]
    n = len(z)
    dz = z[1]-z[0]
    
    D2 = second_derivative_matrix(n,1) #Matrix of second differentiation
    A = D2
    A = A.astype('complex')
    
    A[0]   = np.zeros(n)
    A[0,0] = 1
    A[-1]  = np.zeros(n)
    A[-1,-1] = 1
    
    b = np.zeros(n)
    b[1:-1]= dz**2 * rhs[1:-1]
    
    w = np.linalg.solve(A,b)
    w_tot = np.concatenate((w,0*z_[z_>=11e3]))
    return w_tot

def couple_dgw(Zbpa,tau,k,TTref):
    rhoww=0*Zbpa.zz*1j
    rhoww1 = rhoww
    
    c=0.7
    
    ws=[]

    for i in range(80):
        print(i, end = ' ')
        try:
            Zbpa.ww = np.real(c*rhoww+(1-c)*rhoww1)#/Zbpa.rho_
            Zbpa.integrate()
            rhs=-Zbpa.rho_*tau*k**2*g*((Zbpa.TT/TTref)-1)
            rhoww1=rhoww
            rhoww = dgw_mode(Zbpa.zz,rhs)
            ws.append(rhoww)#/Zbpa.rho_)
        except ValueError:
            return ws

    return ws
```

```python
wi = np.array([np.concatenate((np.sin(n*np.pi*Ref.zz[Ref.zz<11e3]/11e3),0.*Ref.zz[Ref.zz>=11e3])) for n in range(1,10)])
def to_minimize(coefs,Zbpa,tau,k,TTref):
    ww = np.dot(coefs,wi)
    z = Zbpa.zz[Zbpa.zz<11e3]
    rhow = (ww)[Zbpa.zz<11e3]
    n = len(z)
    dz = z[1]-z[0]
    
    A = second_derivative_matrix(n,1)/dz**2 #Matrix of second differentiation
    
    d2rhowdz2 = np.dot(A,rhow)
    Zbpa.ww=ww
    Zbpa.integrate()

    return 1e15*np.sum((d2rhowdz2 + tau*k**2*g*((Zbpa.TT/TTref)-1)[Zbpa.zz<11e3])**2)

# Simple gradient descent
def estimate_gradient(coefs,Zbpa,tau,k,TTref):
    eps = 1e-6
    bases = np.eye(len(coefs))
    ref = to_minimize(coefs,Zbpa,tau,k,TTref)
    return np.array([-(to_minimize(coefs-eps*bases[i],Zbpa,tau,k,TTref)-ref)/eps for i in range(len(wi))])
```

```python
coefs = np.zeros(len(wi))

Ref1 = ZPB(zz,ww,Tref=244.8)
Ref1.integrate()

ws = []
ws.append(np.dot(coefs,wi))
alpha = 1e-8
alphas = alpha*2**np.arange(4)

for i in range(10):
    print(i,to_minimize(coefs,Ref1,86400,1e-6,Ref.TT))
    grad = estimate_gradient(coefs,Ref1,86400,1e-6,Ref.TT)
    #print(grad)
    best = np.argmin([to_minimize(coefs - a*grad,Ref1,86400,1e-6,Ref.TT) for a in alphas])
    coefs = coefs - alphas[best]*grad
    ws.append(np.dot(coefs,wi))
```
